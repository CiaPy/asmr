```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from matplotlib.animation import FuncAnimation
```

```python
friend= "Friends are the companions in life that every soul craves for. We get a family by born but we make a family by choice through making friends. But the course of life often makes the distance between you and your friends whether for studies, job or even death. This is up to you if you let the distance wipe out the warmth of your friendship or keep your friends closer by sending your heartfelt words. For you, we assembled a bunch of messages which you can send to your friends or dedicate him/her to tell how much you miss them or express your emotion by sending long distance friendship message."
#friend = "Vous êtes trop généreuse pour vous jouer de moi ; si vos sentiments sont encore ce qu’ils étaient au mois d’avril dernier, dites-le-moi franchement ; mes désirs, mes affections n’ont point changé, mais un mot de vous les forcera pour jamais au silence. Nom de Dieu. Tout bouillonne à l’intérieur de moi.  Les mots ne suffisent plus.  J’ai envie de mordre dans les choses, avec mes dents.  Je t’adore. Tu me fais croire que tout est possible. Je pense à toi tout le temps. Je pense à toi le matin, en marchant dans le froid. Je fais exprès de marcher lentement pour pouvoir penser à toi plus longtemps. Je pense à toi le soir, quand tu me manques au milieu des fêtes, où je me saoule pour penser à autre chose qu'à toi, avec l'effet contraire. Je pense à toi quand je te vois et aussi quand je ne te vois pas. J'aimerais tant faire autre chose que penser à toi mais je n'y arrive pas. Si tu connais un truc pour t'oublier, fais le moi savoir."
wordcloud = WordCloud().generate(friend)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
# png image
image_file = "cloud.png"

# create mask
cloud_mask = np.array(Image.open(image_file))

# generate word cloud
wc = WordCloud(font_path= "Milk Peach.ttf",   mode = "RGBA", background_color="#fff9ec", mask=cloud_mask, colormap='Pastel2')
wc.generate(friend)
# Convert the word cloud to an image
corrected_image = wc.to_image()
#corrected_image = wc_image.rotate(180)
#corrected_image.to_file('wc1.png')

# plot the word cloud
# plt.figure(figsize=(8,6), dpi=120)
# plt.imshow(wc, interpolation='bilinear')
# plt.axis("off")

# plt.savefig('wc_final_firend.png', dpi=250)
```

```python
# Set up the figure and axis
fig, ax = plt.subplots(figsize=(8,6), dpi=120)
fig.patch.set_facecolor('#fff9ec')  # Set the background color for the figure
ax.set_facecolor('#fff9ec')  # Set the background color for the axes

ax.imshow(corrected_image, interpolation='bilinear')
ax.axis("off")

# Function to animate
def update(frame):
    ax.clear()
    ax.set_facecolor('#fff9ec')  # Ensure the background color persists through frames
    ax.imshow(corrected_image, interpolation='bilinear')
    ax.axis("off")
    ax.set_xlim(frame, corrected_image.width - frame)
    ax.set_ylim(corrected_image.height - frame, frame)

# Create animation
ani = FuncAnimation(fig, update, frames=np.arange(0, corrected_image.width//2, 5), interval=50)

# Save the animation as a GIF
ani.save('wc_zoom_animation.gif', writer='imagemagick')

plt.show()
```
